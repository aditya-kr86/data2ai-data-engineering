## 1. What is HTTP?

**HTTP** stands for **HyperText Transfer Protocol**. 

Don't let the fancy name intimidate you. A "protocol" is just a strict set of rules for communication. HTTP is simply the agreed-upon grammar that all computers use to ask for and send data over the web.

### The Request-Response Cycle
Every single interaction on the internet is a two-part conversation:
1. **The Request:** The Client (your browser, or your Python code) asks the Server for something.
2. **The Response:** The Server replies with the answer (or an error).

When making a **Request**, the client uses a specific "verb" to tell the server what it wants to do. The most common is **GET** (meaning: "Please give me this data"). Another common one is **POST** (meaning: "Here is some new data, please save it").

## 2. Status Codes: The Server's Mood Ring

When the server sends a **Response** back, it always includes a 3-digit number called a **Status Code**. This code instantly tells the client how the request went, without having to read the whole message.

Here is the cheat sheet for the most common status codes you will see as a Data Engineer:

| Code | Category | Meaning | Real-World Equivalent |
| :--- | :--- | :--- | :--- |
| **200** | Success | **OK.** Everything worked perfectly! | The chef cooked exactly what you ordered. |
| **400** | Client Error | **Bad Request.** You asked for something incorrectly. | You ordered "fried ice", which makes no sense. |
| **403** | Client Error | **Forbidden.** You don't have a password for this. | Trying to enter the VIP lounge without a ticket. |
| **404** | Client Error | **Not Found.** The file doesn't exist here. | Ordering a taco at a sushi restaurant. |
| **500** | Server Error | **Internal Server Error.** The server crashed. | The kitchen caught on fire. Not your fault! |

Let's write some Python code to actually send an HTTP Request and look at the Status Code!

```python
# In Data Engineering, the 'requests' library is the industry standard for HTTP.
# It is much friendlier than the 'urllib' we used in previous lessons!
import requests

print("[Client] Sending a GET request to GitHub's public API...")
response = requests.get("https://api.github.com")

print(f"[Server] Replied with Status Code: {response.status_code}")

if response.status_code == 200:
    print("Success! We got the data.")
else:
    print("Something went wrong!")
```

## 3. HTTP vs. HTTPS (The 'S' stands for Secure)

You may have noticed that most web addresses start with `https://` instead of `http://`. What's the difference?

Standard **HTTP** sends data as plain text. This means if a hacker intercepts your internet traffic at a coffee shop Wi-Fi, they can read everything—including your passwords and credit card numbers.

### The Postcard vs. The Locked Safe

* **HTTP is like a Postcard:** The mailman, the sorting facility, and anyone who glances at it can read the message written on the back.
* **HTTPS is like a Locked Safe:** Before the data leaves your computer, it is heavily encrypted using technology called **SSL/TLS**. It travels across the internet as a scrambled mess. Only the destination server has the key to unlock and read it.

## 4. Why Data Engineers Care

As a Data Engineer, you will constantly write code that extracts data from external services (like pulling sales data from Stripe, or advertising data from Facebook). 

To do this, you will use Python to make **HTTP Requests**.
1. You must know how to construct the Request properly.
2. You must always check the **Status Code** (If it's a 500, your pipeline should pause and try again later!).
3. You must ensure you are using **HTTPS**, because you are handling sensitive company data that cannot be intercepted by hackers.

## 🔑 Key Takeaways

* **HTTP** is the universal language and set of rules computers use to talk to each other on the web.
* Communication happens in a **Request-Response** cycle.
* Servers return **Status Codes** to indicate success (200s), client errors (400s), or server errors (500s).
* **HTTPS** uses SSL/TLS encryption to scramble data, ensuring no one can read it while it travels across the network.
* Data Engineers use Python libraries like `requests` to automate HTTP communication and download data.
