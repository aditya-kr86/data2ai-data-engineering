## 1. What is a REST API?

**REST** stands for *Representational State Transfer*. (You can immediately forget what the acronym stands for—nobody uses the full name!)

Think of a **REST API** as a highly organized librarian managing a giant filing cabinet.

In a REST API, everything is treated as a **Resource**. 
Instead of having confusing, messy URLs like `api.website.com/fetch_user_data_now_please_22`...
A REST API uses clean, noun-based URLs based on the resource you want:
* `api.website.com/users` (The drawer containing all users)
* `api.website.com/users/5` (The specific folder for User #5)
* `api.website.com/posts` (The drawer containing all blog posts)

## 2. CRUD and the Four HTTP Verbs

If the URL tells the server *what* resource you are looking at, how do you tell the server *what action* to take on that resource?

Whenever we interact with data in a database, we only ever do four basic things. We call this **CRUD**:
* **C**reate new data
* **R**ead existing data
* **U**pdate existing data
* **D**elete data

<br>

In a REST API, we map these four CRUD actions directly to **HTTP Methods** (the "verbs" we put in our request):

| CRUD Action | HTTP Method | Example URL | What it does |
| :--- | :--- | :--- | :--- |
| **Read** | `GET` | `/users/5` | "Give me the data for User 5." |
| **Create** | `POST` | `/users` | "Here is new data, create a new user." |
| **Update** | `PUT` | `/users/5` | "Here is updated data, replace User 5." |
| **Delete** | `DELETE` | `/users/5` | "Destroy User 5's record." |

## 3. Python Examples: The CRUD Cycle

Let's put this into practice! We are going to use a free, public testing API called **JSONPlaceholder**. It acts like a fake blog database where we can view, create, update, and delete "posts".

### Action 1: READ (GET)
Let's ask the server to read post number 1.

```python
import requests

base_url = "https://jsonplaceholder.typicode.com"

# READ: Use GET to ask for a specific resource
print("[1] Executing GET request...")
response_get = requests.get(f"{base_url}/posts/1")

print(f"Status Code: {response_get.status_code}")
print("Data Received:", response_get.json())
print("-" * 40)
```

### Action 2: CREATE (POST)
Now, let's create a brand new post. Notice that the URL changes (we just use `/posts`, not a specific ID), and we must include a **payload** of data.

```python
# CREATE: Use POST and provide a payload body
new_post_data = {
    "title": "My First Data Engineering Post",
    "body": "Learning REST APIs is fun!",
    "userId": 1
}

print("\n[2] Executing POST request...")
response_post = requests.post(f"{base_url}/posts", json=new_post_data)

# Notice the status code! 201 means "Created Successfully"
print(f"Status Code: {response_post.status_code}")
print("Server Response (confirming creation):", response_post.json())
print("-" * 40)
```

### Action 3: UPDATE (PUT)
Oops, we made a typo in our post, or we want to change it. We know the ID of the post we want to change, so we use `PUT` on that specific URL and send the completely updated payload.

```python
# UPDATE: Use PUT on a specific ID with the new data
updated_post_data = {
    "id": 1,
    "title": "UPDATED TITLE: Data Engineering Rocks",
    "body": "I changed my mind, REST APIs are AWESOME!",
    "userId": 1
}

print("\n[3] Executing PUT request...")
response_put = requests.put(f"{base_url}/posts/1", json=updated_post_data)

print(f"Status Code: {response_put.status_code}")
print("Server Response (confirming update):", response_put.json())
print("-" * 40)
```

### Action 4: DELETE (DELETE)
Finally, we don't want this post anymore. We tell the server to destroy it. We don't need to send a payload, we just point the `DELETE` verb at the specific URL.

```python
# DELETE: Use DELETE on a specific ID
print("\n[4] Executing DELETE request...")
response_delete = requests.delete(f"{base_url}/posts/1")

print(f"Status Code: {response_delete.status_code}")
print("Server Response (Usually empty for delete):", response_delete.json())
```

## 4. Why Data Engineers Care

While Software Engineers building mobile apps will use all four verbs constantly, **Data Engineers have a favorite verb: `GET`.**

A typical Data Engineering task is called **Ingestion**. This means reading large amounts of data from an external system so we can analyze it. 

For example, you might write a Python script that runs every night at midnight, sends a `GET` request to the Salesforce API to download all the new sales made that day, and saves them to your company's analytics database.

However, you will also occasionally use `POST` or `PUT` to send alerts (like posting a message to a Slack channel if your pipeline fails) or to write processed data back to an external system!

## 🔑 Key Takeaways

* **REST** is a popular set of rules for designing APIs, organizing data neatly into "Resources" represented by clean URLs.
* **CRUD** stands for Create, Read, Update, Delete. These are the four fundamental ways we interact with data.
* In REST, CRUD maps perfectly to HTTP verbs:
  * **Read** = `GET`
  * **Create** = `POST`
  * **Update** = `PUT`
  * **Delete** = `DELETE`
* The Python `requests` library handles all of these methods easily (e.g., `requests.get()`, `requests.post()`, etc.).
