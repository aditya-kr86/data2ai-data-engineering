Here is how ordering food maps perfectly to how the internet works:

1. **The Request:** You type `amazon.com` and hit Enter. This is like deciding you want food from "Mario's Pizza".
2. **DNS Lookup:** Your phone needs Mario's exact street address. It checks the phonebook (DNS) to find the address (IP Address).
3. **The Journey:** Your computer sends a digital messenger (an HTTP Request) across the internet's highways to that exact IP Address.
4. **The Web Server (The Kitchen):** The messenger arrives at a remote computer called a Web Server. The server looks at the request, gathers the files needed (HTML, images), and packages them up.
5. **The Response:** The messenger drives back to your laptop and hands your browser the package. Your browser unpacks it and displays the webpage!

## 2. Deep Dive: The DNS Lookup

Let's zoom in on Step 2. As we learned in the last lesson, humans type Domain Names (like `github.com`), but computers only route traffic to IP Addresses (like `140.82.113.3`).

<br>

<img src="../assets/Module_01/01_05_02.png" alt="A magnifying glass looking at a giant digital phonebook labeled 'DNS Resolver', translating the text 'github.com' into the numbers '140.82.113.3'." width="75%">
</div>
<br>

When you press Enter, your browser actually does a quick multi-step check:
1. **Browser Cache:** Have I visited this site recently? Do I already remember the IP address? If no...
2. **OS Cache:** It asks your Operating System (Windows/Mac/Linux), "Do *you* know the IP address?" If no...
3. **DNS Resolver:** Your computer sends a message to a dedicated DNS server on the internet saying, "Help! What is the IP for github.com?"

Once the DNS server replies with the IP Address, the real journey can begin.

```python
# Let's use Python to do exactly what your browser does in Step 2!
import socket

website_url = "www.wikipedia.org"

print(f"[Browser] Asking DNS for the IP address of {website_url}...")

# This single line reaches out to the DNS system
website_ip = socket.gethostbyname(website_url)

print(f"[DNS] Found it! The IP address is: {website_ip}")
print(f"[Browser] Great, now I can send my request directly to {website_ip}!")
```

## 3. What is a Web Server?

In Step 4, your request arrives at the destination. But what is waiting there to receive it?

A **Web Server** is just a normal computer (usually running Linux, as we learned in Lesson 1.2). But it is running special software that does one thing: **It listens.**

It sits there 24 hours a day, 7 days a week, waiting for digital messengers to arrive. 

<br>
<div align="center">
<img src="../assets/Module_01/01_05_03.png" alt="A friendly robot waiter labeled 'Web Server Software' standing in front of a giant computer, handing a tray of files (HTML, CSS) to a digital messenger." width="75%">
</div>
<br>

When a request arrives, the Web Server says:
1. *"What do you want? Ah, the homepage."*
2. It grabs the HTML file for the homepage from its File System.
3. It attaches a "Status Code" to the package. (A **200** means "Success, here is your stuff!". A **404** means "Error, I couldn't find that file!").
4. It sends the package back to you.

```python
# Let's act like a web browser and talk to a real web server!
import urllib.request

url = "http://www.google.com"

print(f"Sending a request to the web server at {url}...")
response = urllib.request.urlopen(url)

# Let's check the "Status Code" the server sent back
status_code = response.getcode()

if status_code == 200:
    print("\nServer replied: '200 OK'! (Success)")
    print("The server sent back a package of data. Here are the first 150 characters:\n")
    
    # Read the data the server handed to us
    html_content = response.read().decode('utf-8')
    print(html_content[:150] + "...")
else:
    print(f"Uh oh, the server replied with status code: {status_code}")
```

## 4. Why Data Engineers Care

When you use a browser, the Web Server sends back HTML code to make a pretty webpage for a human to read.

But Web Servers don't *have* to send back pretty webpages. They can also send back raw, pure data (like lists of numbers, customer records, or sensor readings). When a Web Server is set up to send raw data to other computers instead of human webpages, we call it an **API** (Application Programming Interface).

As a Data Engineer, you will write Python scripts that act exactly like web browsers. But instead of downloading webpages to read, your scripts will request raw data from APIs, download it, and save it to your databases!

## 🔑 Key Takeaways

* The internet journey is: **URL → DNS Lookup → IP Address → Request sent to Web Server → Response received**.
* **DNS** translates human-friendly URLs into computer-friendly IP addresses.
* A **Web Server** is a computer running software that listens for requests and serves files or data back.
* Web Servers send back **Status Codes** to tell you if the request was successful (e.g., 200 OK) or failed (e.g., 404 Not Found).
* Data Engineers use this exact same mechanism to request raw data from remote servers (APIs).
