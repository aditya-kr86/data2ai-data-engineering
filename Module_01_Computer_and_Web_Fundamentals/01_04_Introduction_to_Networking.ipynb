## 1. What is a Network?

At its simplest, a **network** is just two or more computers connected together so they can share data. 

If you connect your laptop to your printer using a cable, that is a tiny network. If you connect millions of computers, servers, and smartphones all over the world, that is the biggest network of all: **The Internet**.

## 2. IP Addresses: How Computers Find Each Other

If you want to send a physical letter to a friend, you can't just write "To my friend" on the envelope. The postal service needs a specific, unique **House Address** (like 123 Main St).

Computers work the exact same way. Every single device connected to a network gets a unique numerical address called an **IP Address** (Internet Protocol Address).

An IP address usually looks like four numbers separated by dots, like this: `192.168.1.5` or `142.250.190.46`.

When your phone (Frontend) wants to ask Instagram (Backend) for a new picture, your phone sends a request across the internet directly to Instagram's IP Address!

## 3. Domain Names and DNS: The Phonebook

There's a problem with IP addresses: **Humans are terrible at remembering numbers.** 

Imagine if, to watch a video, you had to type `142.251.46.238` into your browser instead of `youtube.com`. It would be a nightmare!

To solve this, we invented **Domain Names** (like `google.com`, `amazon.com`). 

But computers only understand IP Addresses. So, we need a translator. This is called the **DNS** (Domain Name System). 

### The Phonebook Analogy
DNS is literally the internet's Contacts App or Phonebook. 
1. You open your phone and tap "Mom" (Domain Name).
2. Your phone looks up "Mom" in your contacts (DNS).
3. It finds her actual phone number `555-1234` (IP Address) and dials it behind the scenes.

<br>

<br>

Let's write a tiny piece of Python code to act like DNS and look up the IP address of a famous website!

```python
# We import the 'socket' tool, which helps Python talk to the network
import socket

# 1. This is the Domain Name (easy for humans)
domain_name = "www.google.com"

# 2. We ask the DNS phonebook to get the actual IP Address
ip_address = socket.gethostbyname(domain_name)

# 3. Print the result!
print(f"The domain name '{domain_name}' actually lives at IP Address: {ip_address}")
```

## 4. Why Data Engineers Care About Networking

You might be thinking, "I want to engineer data, not set up Wi-Fi routers!"

As a Data Engineer, the data you need to work with almost *never* lives on your local laptop. It lives on a remote Database server somewhere in the cloud.

To connect your Python script to that Database, you will need to give your script the **IP Address** (or Domain Name) of the database, so your code knows exactly what "house" on the internet to drive to in order to pick up the data.

Let's simulate our code traveling across the network to grab some real data from a server.

```python
# We use 'urllib' to simulate opening a URL over the network
import urllib.request

print("Starting our car and driving across the network...\n")

# Send a request to an example website
response = urllib.request.urlopen("http://www.example.com")
html_data = response.read().decode('utf-8')

print(f"Success! We connected to the server and downloaded {len(html_data)} characters of data.")
print("Here is a tiny sneak peek of the raw data we brought back:")
print(html_data[:100] + "...")
```

## 🔑 Key Takeaways

* A **Network** is a group of computers connected together to share data. The Internet is the biggest network.
* An **IP Address** is a unique number assigned to every computer on a network, acting exactly like a house's mailing address.
* A **Domain Name** (like `google.com`) is a human-readable name used because humans are bad at remembering numbers.
* **DNS** acts like a giant phonebook, translating easy Domain Names into the exact IP Addresses computers need to connect.
* Data Engineers rely on networking to connect their data pipelines to remote databases.
