# What is Apache Spark?

Apache Spark is an open-source distributed data processing engine designed for large-scale data processing.

It allows us to process data across multiple machines in parallel.

Spark is widely used for:

- Data Engineering
- Data Analytics
- Machine Learning
- Streaming
- AI Data Pipelines

Created at:
UC Berkeley (2010)

Current Maintainer:
Apache Software Foundation

# Why Do We Need Spark?

Imagine we have:

- 1 TB sales dataset

Can a single laptop process this efficiently?

❌ No

Problems:

- Limited RAM
- Limited CPU
- Slow execution

We need distributed computing.

# Single Machine Processing

Traditional systems process everything on a single computer.

Example:

Dataset → One Machine → Result

As data grows:

- Memory becomes insufficient
- Processing becomes slower
- Scaling becomes difficult

<img src="../assets/Module_10/10_01_02.png" alt="Single machine vs cluster processing" width="75%">

# Distributed Computing

Instead of one machine doing all work:

Data is split across multiple machines.

Each machine processes part of the data.

Results are combined together.

Benefits:

- Faster processing
- Better scalability
- Fault tolerance
- Lower execution time

# What Makes Spark Special?

Before Spark, Hadoop MapReduce was the dominant Big Data framework.

MapReduce worked well but had limitations:

- Heavy disk usage
- Slow iterative processing
- Complex development

Spark introduced:

- In-memory processing
- Faster execution
- Simpler APIs
- Unified ecosystem

# Spark vs Hadoop

| Feature | Hadoop MapReduce | Spark |
|----------|-----------------|--------|
| Processing Speed | Slower | Faster |
| Memory Usage | Disk-Based | In-Memory |
| Development | Complex | Easier |
| Machine Learning | Limited | Built-In Support |
| Streaming | Limited | Native Support |
| Iterative Workloads | Slow | Fast |

<img src="../assets/Module_10/10_01_03.png" alt="Hadoop MapReduce vs Apache Spark comparison" width="75%">

# Why Spark is Faster

### MapReduce:

Read Disk  
    ↓  
Process  
    ↓  
Write Disk  
    ↓  
Read Disk Again  
    ↓  
Process

### Spark:

Read Once  
    ↓  
Store In Memory  
    ↓  
Multiple Operations

Less disk I/O = Faster execution

# Spark Ecosystem

# Spark Core

Spark Core is the foundation of Spark.

Responsibilities:

- Task Scheduling
- Memory Management
- Fault Recovery
- Distributed Execution

Every Spark application uses Spark Core.

# Spark SQL

Spark SQL allows working with structured data.

Examples:

- CSV
- JSON
- Parquet
- Database Tables

Features:

- SQL Queries
- DataFrames
- Optimized Execution

# Structured Streaming

Used for real-time processing.

Examples:

- Live Transactions
- IoT Sensors
- Website Clickstreams
- Kafka Streams

Process data continuously as it arrives.

# MLlib

Spark's Machine Learning Library.

Supports:

- Classification
- Regression
- Clustering

Allows machine learning at scale.

# GraphX

Graph Processing Framework.
